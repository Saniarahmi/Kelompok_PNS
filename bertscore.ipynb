```python
!pip install python-dotenv google-generativeai

import requests
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Memuat variabel lingkungan dari file .env
load_dotenv()

API_KEY = os.getenv('GEMINI_API_KEY')

genai.configure(
    api_key=API_KEY
)

model = genai.GenerativeModel('gemini-pro')
chat = model.start_chat(history=[])
instruction = "Sekarang anda berperan sebagai chatbot wisata Raja Ampat. Dalam percakapan ini, berikan jawaban tentang destinasi wisata Raja Ampat, seolah-olah Anda menjelaskannya kepada turis atau wisatawan."

while True:
    question = input("Kamu: ")

    if "raja" in question.lower() and "ampat" in question.lower():
        # Menerapkan filter berdasarkan kata kunci yang relevan dengan Raja Ampat
        response_text = chat.send_message(instruction + question).text
        response = response_text
    else:
        response_text = chat.send_message(instruction + question).text
        response = response_text

    print('\n')
    print(f"Bot: {response}")
    print('\n')
    instruction = ''
```

```python
!pip install bert-score

from bert_score import score

# Ganti dengan list prediksi dan jawaban yang benar
predicted_answers = ["Ini adalah contoh jawaban prediksi.", "Jawaban lain di sini."]
true_answers = ["Ini adalah jawaban yang benar.", "Jawaban lain yang benar."]

# Hitung skor BERT
P, R, F1 = score(predicted_answers, true_answers, lang="id", verbose=True)

# Cetak skor
print(f"Precision: {P.mean()}")
print(f"Recall: {R.mean()}")
print(f"F1 Score: {F1.mean()}")
```
